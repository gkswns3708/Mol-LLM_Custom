```python
import torch
import os

# 1.   (     )
ckpt_path = "/home/jovyan/CHJ/Mol-LLM_Custom/checkpoint/Custom_LLaDA/stage1_llm_pretraining/epoch=00-step=1.ckpt"
checkpoint = torch.load(ckpt_path, map_location='cpu')

# 2. state_dict 
state_dict = checkpoint['state_dict']

# 3.   
total_params = 0
total_bytes = 0

print(f"{'Layer Name':<60} | {'Shape':<20} | {'Dtype'}")
print("-" * 100)

for key, tensor in state_dict.items():
    # LoRA    (  )
    if "lora" in key:
        #   (: 4096 * 64)
        num_params = tensor.numel()
        #     (float32=4, float16=2, bfloat16=2)
        elem_size = tensor.element_size()
        
        total_params += num_params
        total_bytes += num_params * elem_size
        
        # ()  3  
        if total_params < 200000: 
            print(f"{key[:55]:<60} | {str(list(tensor.shape)):<20} | {tensor.dtype}")

print("-" * 100)
print(f"  ")
print(f"1.   : {total_params:,} ")
print(f"2.   : {total_bytes / (1024**2):.2f} MB")
print(f"3.       ( ):  {total_bytes / (1024**2) + 1:.2f} MB")
```
