```python
import pandas as pd
from datasets import load_from_disk

# 1. 데이터셋 로드 (경로는 사용자 환경에 맞게 설정됨)
path_root = "Mol-LLM_Custom/dataset/real_train/"
writer_dataset = load_from_disk(path_root + "mistralai-Mistral-7B-Instruct-v0.3_string+graph_q32_test_3.3M_0415_writer")
download_original_dataset = load_from_disk(path_root + "mistralai-Mistral-7B-Instruct-v0.3_string+graph_q32_test_3.3M_0415_download_original")
download_preprocessed_dataset = load_from_disk(path_root + "mistralai-Mistral-7B-Instruct-v0.3_string+graph_q32_test_3.3M_0415_download_preprocessed")

# 2. 각 데이터셋의 unique task 추출 (Set으로 변환하여 연산 속도 최적화)
# pd.Series를 사용하여 빠르게 유니크 값을 뽑고, 집합 연산을 위해 set으로 변환합니다.
tasks_writer = set(pd.Series(writer_dataset['task']).unique())
tasks_original = set(pd.Series(download_original_dataset['task']).unique())
tasks_preprocessed = set(pd.Series(download_preprocessed_dataset['task']).unique())

# 3. 비교 및 표 생성 함수 정의
def compare_datasets(base_name, base_set, target_name, target_set):
    # 집합 연산
    common = base_set & target_set              # 교집합 (공통)
    only_base = base_set - target_set           # 차집합 (기준 데이터셋에만 있음)
    only_target = target_set - base_set         # 차집합 (비교 대상에만 있음)

    # 결과 데이터프레임 생성
    df_compare = pd.DataFrame([
        {
            '구분': '공통 (Intersection)',
            '개수': len(common),
            'Task 목록': list(common) if common else "-"
        },
        {
            '구분': f'{base_name} 에만 있음',
            '개수': len(only_base),
            'Task 목록': list(only_base) if only_base else "-"
        },
        {
            '구분': f'{target_name} 에만 있음',
            '개수': len(only_target),
            'Task 목록': list(only_target) if only_target else "-"
        }
    ])
    
    print(f"\n{'='*20} [ {base_name} vs {target_name} ] 비교 결과 {'='*20}")
    # 보기 좋게 출력 (Jupyter 환경이면 display(df_compare) 사용 가능)
    print(df_compare.to_string(index=False)) 
    return df_compare

# 4. 비교 실행

# Case 1: Writer vs Download Original
df1 = compare_datasets('Writer', tasks_writer, 'Original', tasks_original)

# Case 2: Writer vs Download Preprocessed
df2 = compare_datasets('Writer', tasks_writer, 'Preprocessed', tasks_preprocessed)
```
