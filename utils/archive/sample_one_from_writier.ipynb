```python
import os
from datasets import load_from_disk

# 1. 원본 데이터셋 경로 및 저장할 경로 설정
input_path = "Mol-LLM_Custom/dataset/real_train/mistralai-Mistral-7B-Instruct-v0.3_string+graph_q32_test_3.3M_0415_wrtier_100_sampled"
output_path = "Mol-LLM_Custom/dataset/real_train/mistralai-Mistral-7B-Instruct-v0.3_string+graph_q32_test_3.3M_0415_writier_one"  # 저장할 경로 (원하는 경로로 수정하세요)

# 2. 데이터셋 로드
print("데이터셋을 로드 중입니다...")
writier_testset = load_from_disk(input_path)

# 3. 첫 번째 샘플([0])만 추출
# 주의: testset[0]은 dict를 반환하므로, Dataset 객체 유지를 위해 select()를 사용합니다.
writier_testset = writier_testset.filter(lambda x: x['task'] == 'bace').filter()
writier_single_sample_dataset = writier_testset.select([1])

# 4. 추출한 샘플을 디스크에 저장
print(f"첫 번째 샘플을 {output_path}에 저장 중입니다...")
writier_single_sample_dataset.save_to_disk(output_path)

print("저장이 완료되었습니다.")

# 확인을 위해 저장된 데이터 다시 로드해보기 (선택 사항)
check_dataset = load_from_disk(output_path)
print(f"저장된 데이터셋 개수: {len(check_dataset)}")
print(f"데이터 내용: {check_dataset[0]}")
```
